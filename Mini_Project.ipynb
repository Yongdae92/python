```python
import random

words = ["apple", "banana", "orange", "grape", "lemon"]
chosen_word = words[random.randint(0, len(words)-1)]  # ê³¼ì¼ì„ ë¬´ì‘ìœ„ë¡œ ì„ íƒí•©ë‹ˆë‹¤.
length = len(chosen_word)  # ì„ íƒí•œ ë‹¨ì–´ì˜ ê¸¸ì´ë¥¼ êµ¬í•©ë‹ˆë‹¤.
answer = ["_"] * length  # ë‹¨ì–´ì˜ ê¸¸ì´ë§Œí¼ ë°‘ì¤„ì„ ìƒì„±í•©ë‹ˆë‹¤.
attempts = 10  # í”Œë ˆì´ì–´ì˜ ì‹œë„ íšŸìˆ˜ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.

def display_word():
    print(" ".join(answer))  # í˜„ì¬ê¹Œì§€ ë§ì¶˜ ê¸€ìë“¤ì„ ì¶œë ¥í•©ë‹ˆë‹¤. ê¸°ì¡´ì— ìˆë˜ "-"ë¥¼ ë§ì¶˜ ë‹¨ì–´ë¡œ ëŒ€ì²´í•©ë‹ˆë‹¤.

def play_game():
    global attempts  # attempts ë³€ìˆ˜ë¥¼ ì „ì—­ ë³€ìˆ˜ë¡œ ì„ ì–¸í•©ë‹ˆë‹¤.
    while attempts > 0:
        display_word()
        guess = input("ì•ŒíŒŒë²³ì„ ì…ë ¥í•˜ì„¸ìš”: ")
        if guess in chosen_word:
            print("ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨êµ°ìš”. ë” í˜ë‚´ë´ìš”.")
            for i in range(length):
                if chosen_word[i] == guess:
                    answer[i] = guess
        else:
            attempts -= 1
            print("í‹€ë ¸ìŠµë‹ˆë‹¤. ë‚¨ì€ ì‹œë„ íšŸìˆ˜:", attempts)
        if "_" not in answer:
            print("ì¶•í•˜í•©ë‹ˆë‹¤! ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤.")
            break

play_game()
```
